```python
from keras.applications.resnet50 import ResNet50, decode_predictions, preprocess_input
import requests
from keras.preprocessing import image
import numpy as np

from io import BytesIO
from PIL import Image
```

```python
model = ResNet50(weights= 'imagenet')
```

```python
#![](url)
```

![](https://www.rspcasa.org.au/wp-content/uploads/2019/01/Adopt-a-cat-TOC-380x390-fill-crop-constrain-q70-toc_image.jpg)

```python
url_image = 'https://www.rspcasa.org.au/wp-content/uploads/2019/01/Adopt-a-cat-TOC-380x390-fill-crop-constrain-q70-toc_image.jpg'

response  = requests.get(url_image)
img = Image.open(BytesIO(response.content))
img = img.resize((224,224))
img
```

```python
X = image.img_to_array(img)
#Dodanie wymiaru
X = np.expand_dims(X, axis=0)
X.shape
```

```python
y_pred = model.predict(X)
```

```python
y_pred
```

```python
decode_predictions(y_pred, top=3)
```

```python
# Wczytanie danych
orders = pd.read_csv('orders.csv', sep=";", encoding="ISO-8859-1")
orderLines = pd.read_csv('orderlines.csv', sep=";")
products = pd.read_csv('products.csv', sep=";")
customers = pd.read_csv('customers.csv', sep=";")
```

```python
mergetables = orders.merge(orderLines, on="orderid")
mergetables = mergetables.merge(products, on ="productid")
mergetables = mergetables.merge(customers, on ="customerid")
mergetables.head(n=2)
```

```python
datab = mergetables[['productid','orderid','productgroupname','householdid']]
datab.head(n=10)
```

```python
abc=datab.groupby(['householdid','productgroupname']).size().reset_index(name='count')
expectedtab = abc[abc['count'] > 1]
expectedtab
expectedtab2 = expectedtab.sort_values(by = "count", ascending=False)
expectedtab2
```

```python
expectedtab22=expectedtab2.groupby(['productgroupname']).size().reset_index(name='count')
expectedtab22
expectedtab22 = expectedtab22.sort_values(by = "count", ascending=False)
expectedtab22
```

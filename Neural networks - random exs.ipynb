Zaprojektuj sieć neuronową realizującą funkcję logiczną XOR. Ogranicz liczbę warstw sieci do minimum. Wytrenuj sieć maksymalnie dla 1000 epok. Sprawdź dokładność z jaką sieć się uczy (użyj funkcji model.predict(…)).  Sprawdź które z poniższych parametrów modelu dadzą najlepsze efekty. Funkcji aktywacji relu, tanh, sigmoid; funkcji straty mean_squared_error lub binary_crossentropy, optymalizatora adam i metryki binary_accuracy.

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, Dropout, Activation
from tensorflow.keras.optimizers import SGD
import numpy as np 
```

```python
X = np.array([[0,0],[0,1],[1,0],[1,1]])
y = np.array([[0],[1],[1],[0]])
```

```python
model = Sequential()
model.add(Dense(8, input_dim=2))
model.add(Activation('tanh'))
model.add(Dense(1))
model.add(Activation('sigmoid'))
```

```python
sgd = SGD(lr=0.1)
model.compile(loss='binary_crossentropy', optimizer=sgd,metrics=['binary_accuracy'])
```

```python
model.fit(X, y, batch_size=1, nb_epoch=1000)
print(model.predict_proba(X))
```

```python
model2 = Sequential()
model2.add(Dense(8, input_dim=2))
model2.add(Activation('relu'))
model2.add(Dense(1))
model2.add(Activation('sigmoid'))
```

```python
sgd = SGD(lr=0.1)
model2.compile(loss='mean_squared_error', optimizer='adam' ,metrics=['binary_accuracy'])
```

```python
model2.fit(X, y, batch_size=1, nb_epoch=1000)
print(model2.predict_proba(X))
```

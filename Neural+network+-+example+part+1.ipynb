### Zadanie 1

Dla modelu z pliku nn01.py wykonaj: 

 Zainicjuj ziarno generatora liczbą s równą twój rok urodzenia – miesiąc – dzień 

 Utrzymując parametr trenowania sieci batch_size=10, narysuj wykres dokładności sieci w funkcji epoki. 
Epoka zmienia się w zakresie 10 do 150, krok 10.

 Wykonaj wykres dokładności sieci w funkcji batch_size, parametr ten przyjmuje wartości 2, 10, 20, 50, 100. Przyjmij liczbę epok równą 100. 

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
import numpy as np
import tensorflow as tf 
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
seed= 19951111
numpy.random.seed(seed)
```

```python
#ladowanie danych
dataset = numpy.loadtxt("dane.csv", delimiter=",")
# split into input (X) and output (Y) variables
X = dataset[:,0:8]
Y = dataset[:,8]
```

```python
#  model
model = Sequential()
model.add(Dense(12, input_dim=8, activation='relu'))
model.add(Dense(8, activation='relu'))
model.add(Dense(1, activation='sigmoid'))
```

```python
#optymalizacja
model.compile(loss='binary_crossentropy', optimizer='adam', metrics=['accuracy'])
```

```python
fittedModel = model.fit(X, Y, epochs=150, batch_size=10)
```

```python
# ewaluacja
scores = model.evaluate(X, Y)
print("\n%s: %.2f%%" % (model.metrics_names[1], scores[1]*100))
```

```python
print(fittedModel.history.keys())
# summarize history for accuracy
```

```python
plt.plot(fittedModel.history['accuracy'])

plt.title('Wykres dokładności sieci')
plt.legend(['train'], loc='upper left')
plt.ylabel('dokładność')
plt.xlabel('epoka')
plt.xlim([10,150])
plt.xticks(np.arange(10, 151, step=10))

plt.show()
```

```python
modelDifferenceBatch2 = model.fit(X, Y, epochs=100, batch_size=2)
modelDifferenceBatch10 = model.fit(X, Y, epochs=100, batch_size=10 )
modelDifferenceBatch20 = model.fit(X, Y, epochs=100, batch_size=20 )
modelDifferenceBatch50 = model.fit(X, Y, epochs=100, batch_size=20 )
modelDifferenceBatch100 = model.fit(X, Y, epochs=100, batch_size=100 )
```

```python
plt.plot(modelDifferenceBatch2.history['accuracy'])
plt.plot(modelDifferenceBatch10.history['accuracy'])
plt.plot(modelDifferenceBatch20.history['accuracy'])
plt.plot(modelDifferenceBatch50.history['accuracy'])
plt.plot(modelDifferenceBatch100.history['accuracy'])
plt.title('Wykres dokładności sieci w zależności od rozmiaru batch_size')
plt.ylabel('dokładność')
plt.xlabel('epoka')
plt.legend(['batch_size=2','batch_size=10','batch_size=20','batch_size=50','batch_size=100'], loc='lower right')
```

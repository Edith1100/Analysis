```python
# Wczytanie danych
orderLines = pd.read_csv('orderlines.csv', sep=";")
orderLines1 = pd.read_csv('orderlines.csv', sep=";")
orderLines2 = pd.read_csv('orderlines.csv', sep=";")
orderLines3 = pd.read_csv('orderlines.csv', sep=";")
products = pd.read_csv('products.csv', sep=";")
products1 = pd.read_csv('products.csv', sep=";")[['productid', 'productgroupname']]
products2 = pd.read_csv('products.csv', sep=";")[['productid', 'productgroupname']]
products3 = pd.read_csv('products.csv', sep=";")[['productid', 'productgroupname']]
```

```python
orderLinesCJ = orderLines1.merge(orderLines2, on = 'orderid', suffixes=['_x','_y']).merge(orderLines3, on = 'orderid').rename(columns = {'productid': 'productid_z'})
orderLinesCJ.head(n=10)
```

```python
differentProducts = orderLinesCJ[(orderLinesCJ.productid_x < orderLinesCJ.productid_y) & (orderLinesCJ.productid_y < orderLinesCJ.productid_z)]
differentProducts.head(n=5)
```

```python
pairs = differentProducts.groupby(["productid_x", "productid_y","productid_z"]);
withCount = pairs.orderid.nunique().reset_index(name = "numberOfOrder")
withCount.head(n=5)
```

```python
products1["productid_x"] = products1.productid
products2["productid_y"] = products2.productid
products3["productid_z"] = products3.productid
```

```python
#Usunięcie zdublowanych kolumn, powstałych podczas mergowania
datab = withCount[['productid_x','productid_y','productid_z','numberOfOrder']]
datab = datab.loc[:,~datab.columns.duplicated()]
datab.head(n=10)
```

```python
withCount = datab.merge(products1, on="productid_x")
withCount = withCount.merge(products2, on="productid_y")
withCount = withCount.merge(products3, on="productid_z")
datac = withCount.sort_values(by = "numberOfOrder", ascending=False)
datac.head(n=10)
```

```python
#Usunięcie zdublowanych kolumn, powstałych podczas mergowania
datad = datac[['productid_x','productid_y','productid_z','numberOfOrder','productgroupname_x','productgroupname_y','productgroupname']]
datad = datad.loc[:,~datad.columns.duplicated()]
datad.rename(columns={'productgroupname': 'productgroupname_z'}, inplace=True)
datad.head(n=10)
```
